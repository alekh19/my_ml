```python
df['A']=df['A'].fillna(1)
```

```python
df['B']=df['B'].fillna(0)
```

```python
df['C']=df['C'].fillna(0)
```

```python
df['E']=df['E'].fillna(0)
```

```python
df['Z']=df['Z'].fillna(0)
```

```python
df['D']=df['D'].fillna(0)
```

```python
df['Number_of_prev_cond']=df['Number_of_prev_cond'].fillna(1)
```

```python
df['Patient_Smoker'].unique()
```

```python
df['Patient_Smoker']=df['Patient_Smoker'].map({'YES':1,'NO':0})
df['Patient_Rural_Urban']=df['Patient_Rural_Urban'].map({'URBAN':1,'RURAL':0})
```

```python
df.info()
```

```python
df.columns
```

```python
col=df[['ID_Patient_Care_Situation', 'Diagnosed_Condition','Patient_Age', 'Patient_Body_Mass_Index','Number_of_prev_cond']]
```

```python
normalized_col=(col-col.mean())/col.std()
df1 = df.drop(['ID_Patient_Care_Situation', 'Diagnosed_Condition','Patient_Age', 'Patient_Body_Mass_Index','Number_of_prev_cond'], 1)
df1 = pd.concat([df1,normalized_col],axis=1)
df1.head()
```

```python
X=df1.drop(['Survived_1_year','Patient_ID'],axis=1)
y=df1['Survived_1_year']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=45)
```

```python
from sklearn import metrics
```

```python
from xgboost import XGBClassifier
```

```python
from sklearn.metrics import f1_score

def f1_eval(y_pred, dtrain):
    y_true = dtrain.get_label()
    err = f1_score(y_true, np.round(y_pred))
    return 'f1_err', err
```

```python
model = XGBClassifier(silent=False, 
                      scale_pos_weight=1,
                      learning_rate=0.01,  
                      colsample_bytree = 0.4,
                      subsample = 1,
                      objective='binary:logistic', 
                      n_estimators=2850, 
                      reg_alpha = 0.3,
                      max_depth=5, 
                      gamma=0)
```

```python
eval_metric = f1_eval
```

```python
model.fit(X_train, y_train, eval_metric=eval_metric, verbose=True)
```

```python
preds = model.predict(X_test)
```

```python
metrics.f1_score(preds,y_test)
```

```python
model.fit(X,y, eval_metric=eval_metric, verbose=True)
```
